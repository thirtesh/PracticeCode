```python
import sklearn as sk
import pandas as pd
import numpy as np
import matplotlib.pyplot as pp
```

```python
X,y=sk.datasets.load_iris(return_X_y=True)
X_train,X_test,y_train,y_test=sk.model_selection.train_test_split(X,y)
```

```python
display(pd.DataFrame(X_train).head())
display(pd.DataFrame(y_train).head())
display(pd.DataFrame(X_train).isnull().sum())
pp.scatter(X_train[:,0],y_train)
```

```python
scaler=sk.preprocessing.StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
display(pd.DataFrame(X_train_scaled).head())
pp.scatter(X_train_scaled[:,0],y_train)
```

```python
quantile_trans=sk.preprocessing.QuantileTransformer(n_quantiles=112,random_state=77)
X_train_scaled = quantile_trans.fit_transform(X_train)
display(pd.DataFrame(X_train_scaled).head())
pp.scatter(X_train_scaled[:,0],y_train)
```
