```python
import numpy as np
```

```python
A=np.random.randint(0,20,[3,3])
B=np.random.randint(0,20,[3,3])
print(A,end='\n\n\n')
print(B)
```

```python
np.dot(A,B)
```

```python
np.matmul(A,B)#same as np.dot()
```

```python
A1=np.linalg.inv(A)
print(A1,end='\n\n\n')
Z=A@A1
print(np.round(Z,10),end='\n\n\n')
```

```python
s=np.sum(A)
v=np.var(A)
print(s,end='\n\n')
print(v,end='\n')
```

```python
mx=np.max(A)
mn=A.min()
print(mx,mn,sep='\n')
```

```python
A5=np.where(A<5,5,A)
print(A5,end='\n\n\n')
Abool=np.where(A>5)
print(Abool)
```

```python
print(np.any(A==0))

print(np.any(A==0,axis=0))
print(np.any(A==0,axis=1))
```

```python
print(np.all(A<15))
print(np.allclose(Z,np.eye(3)))
```
