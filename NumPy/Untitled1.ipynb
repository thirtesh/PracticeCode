```python
a=np.random.rand(3,3)
b=np.random.randint(5,500,[2,3])
c=np.random.randn(2,3)

print('a\n',a,end='\n\n\n')
print('b\n',b,end='\n\n\n')
print('c\n',c,end='\n\n\n')

print(np.concatenate([a,b,c], axis=0))
```

```python
print('transpose of matrix b is\n',b.T,end='\n\n')#transpose
print('matrix AxB is\n',b*c,end='\n\n')

if np.all(b.T==np.transpose(b)):
    print("Note:\'T attribute is the same as numpy transpose function\'")
```
