```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('~/Downloads/titanic.csv')
df.head()
```

```python
peter_data=pd.DataFrame({'Age':[24],'Name':['Peter'],'Survived':['Y']})
pd.concat([df,peter_data], ignore_index=True)
```

```python
df=df.drop(df.index[-1],axis=0)
df.tail()
```

```python
df=df.sort_values(by='Age', ascending=False)
df.head()
```

```python
sno=[i+1 for i in range(50)]
marks=np.random.randint(50,size=50)
marks_df=pd.DataFrame({'Sno':sno,'Marks':marks})
marks_df.head()
```

```python
sno=[i+1 for i in range(50)]
age=np.random.randint(50,size=50)
age_df=pd.DataFrame({'Sno':sno,'Age':age})
age_df.head()
```

```python
#merging two dataframes
cross_df=pd.merge(marks_df,age_df,how='cross')
display(cross_df.head())
display(cross_df.tail())
```

```python
inner_df=pd.merge(marks_df,age_df,how='inner')
display(inner_df.head())
display(inner_df.tail())
```

```python
grouped_df=df.groupby(by='Pclass')
grouped_df.get_group(1).head()
```

```python
#Average age in groups
print(grouped_df['Age'].mean(),
      grouped_df['Age'].min(),
      grouped_df['Age'].count(),   sep='\n\n')
```
