```python
import pandas as pd
```

```python
data=pd.read_csv("~/Downloads/titanic.csv")
df=data.copy()#we copy it to other vairable because we dont wanna change the true values stored in the variable
df.head()
```

```python
df.isnull().sum()
```

```python
print(df['Age'].isnull().sum())
df.fillna({'Age':df['Age'].mean()},inplace=True)
print(df['Age'].isnull().sum())
```

```python
df['Sex']=df['Sex'].map({'male':'0','female':'1'})
df.head()
```

```python
df['lastName']=df['Name'].apply(lambda x:x.split(',')[0])
df['firstName']=df['Name'].apply(lambda x:x.split(',')[1])
df.head()
```

```python
df=df.drop(['PassengerId'], axis=1)
df.head()
```

```python
df0=df.rename(columns={'Sex':'Gender', 'Name':'FullName', 'lastName':'SurName'})
df0.head()
```

```python
df_third_class=df[df['Pclass']==3]
df_third_class.head()
```

```python
df_senior_women=df[(df['Age']>=60) & (df['Sex']=='1')]
df_senior_women.head()
```

```python
#since most of the values are useless, we can just select required columns
df1=df[['Age','Pclass', 'Sex', 'Survived']]
df1.head()
```
